```python
# importing all the libraries 
import seaborn as sns  # ref Seaborn
import pandas as pd # ref  Pandas
import numpy as  np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# read DataSet
pstore = pd.read_csv("./data/googleplaystore.csv") # read csv
pstore.head(10) # top 10
```

```python
# Create a distribution plot for rating
# sns.displot(pstore.Rating)
sns.histplot(pstore.Rating, color = "red", label = "100% Equities", kde = True, linewidth = 0)
# plt.show()
```

```python
#Change the number of bins # kde 概率密布曲線移除
sns.histplot(pstore.Rating,bins = 20, color = "blue", label = "100% Equities", kde = False, linewidth = 0)
```

```python
#Change a distribution plot for rating
# Adding background to the graph
plt.style.use("dark_background")
sns.histplot(pstore.Rating,bins = 20, color = "g", kde = True, linewidth = 0)
plt.title("Distribution of app ratings", fontsize=20, color ='red')
```

```python
# Analyzing the Content Rating column
pstore ['Content Rating'].value_counts() # 對內容rating 行數據清理挖掘 並檢查其中類別
```

```python
# 根據上面樹路 ,由於18歲以上的成年人 未分級的數量比其他的要少很多,將其內容分級中刪除並更新數據及
# Remove the rows with value which are less repressented
pstore = pstore[~pstore['Content Rating'].isin(["Adults only 18+","Unrated"])]

# Resetting the index
pstore.reset_index(inplace = True, drop = True)

pstore['Content Rating'].value_counts() # 對內容rating 行數據清理挖掘 並檢查其中類別
```

```python
plt.style.use("classic")
# Rating 繪製類別餅圖
# Plotting a pie chart
plt.figure(figsize = [9,7])
pstore['Content Rating'].value_counts().plot.pie()
plt.show()
```

```python
# 繪製柱狀圖 Plotting a bar chart
plt.figure(figsize = [9,7])
pstore['Content Rating'].value_counts().plot.barh()
plt.show()
```

```python
# 使用 matplotlib 散點圖
# import all the necessary libraries 
# Plotting the scatter plot
plt.scatter(pstore.Size, pstore.Rating)
plt.show()
```

```python
# Plotting the same thing now using a jointplot  使用Seaborn 散點圖 ,在直方圖,散點圖
# sns.jointplot(pstore.Size, pstore.Rating)
sns.jointplot(pstore.Rating)
plt.show()
```

```python
# Plotting gthe same thing now using a jointplot 回歸曲線
# sns.jointplot(pstore.Size, pstore.Rating, kind ="reg")
sns.jointplot(pstore.Rating, kind ="reg")
plt.show()
```

```python
# Plotting the same thing now using  a jointplot
sns.pairplot(pstore[['Reviews', 'Size', 'Price', 'Rating']])
plt.show()
```
